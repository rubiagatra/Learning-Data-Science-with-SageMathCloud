# Linear Regression Demo

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
print(boston.DESCR)
```

```python
plt.hist(boston.target, bins=50)
plt.xlabel('Harga dalam $1000s')
plt.ylabel('Jumlah rumah')
```

```python
plt.scatter(boston.data[:,5],boston.target)
plt.ylabel('Harga dalam $1000')
plt.xlabel('Jumlah Kamar')
```

```python
boston_df = DataFrame(boston.data)

boston_df.columns = boston.feature_names

boston_df.head()
```

```python
boston_df['Harga'] = boston.target
```

```python
boston_df.head()
```

```python
sns.lmplot('RM','Harga',data=boston_df)
```

# NumPy untuk Univariate Linear Regression

```python
X = boston_df.RM
```

```python
X = np.vstack(boston_df.RM)
```

```python
X
```

```python
X.shape
```

```python
Y = boston_df.Harga
Y
```

```python
#[X 1]
X = np.array( [ [value, 1] for value in X] )
```

```python
X
```

```python
m, b = np.linalg.lstsq(X,Y)[0]
```

```python
plt.plot(boston_df.RM, boston_df.Harga,'o')

x = boston_df.RM

plt.plot(x, m*x + b, 'r', label='Best Fit Line')
```
